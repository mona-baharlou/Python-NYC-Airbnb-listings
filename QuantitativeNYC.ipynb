```python
import pandas as pd
import seaborn as sn
import numpy as ny
import matplotlib.pyplot as plt
```

```python
nyc = pd.read_csv('AB_NYC_2019.csv')
```

```python
nyc
```

```python
plt.hist(nyc['price'])
plt.xlabel('price (in USD Dollars)')
plt.show()
```

```python
plt.hist(nyc['price'], bins = ny.arange(0, 1100, 40))
plt.xlabel('price (in USD Dollars)')
plt.show()
```

```python
plt.scatter(x = nyc['price'] , y = nyc['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.show()
```

```python
plt.scatter(x = nyc['price'] , y = nyc['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0,1100)
plt.show()
```

```python
plt.scatter(x = nyc['price'] , y = nyc['number_of_reviews'], s = 5)
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0,1100)
plt.show()
```
